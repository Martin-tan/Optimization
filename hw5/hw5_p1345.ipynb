```python
# Problem 3
try:
    m = Model("hw5_p3")
    
    # Add Variables
    x=[]
    for i in range(6):      # 1-5 stands for A-E
        tmp=[]
        for j in range(6):
            tmp.append(0)
        x.append(tmp)
    
    for i in range(1,6):
        for j in range(1,6):
            if i!=j:
                x[i][j]= m.addVar(vtype=GRB.INTEGER, name=f'x{i}{j}',lb=0, ub=2) 
    v = m.addVar(vtype=GRB.INTEGER, name="v", lb=0)
    
    m.update()
    m.setObjective(v, GRB.MINIMIZE)
    
    # Add Constraints
    tmp=[]
    m.addConstr(v==sum([x[4][i] for i in range(1,5)]))
    list_=[0,3,3,4,0,5]
    for i in range(1,5):
        if i!=4:
            m.addConstr(v-sum([x[i][j] for j in range(1,5)])>=list_[i])
    for i in range(1,5):
        for j in range(1,5):
            if i!=j:
                m.addConstr(x[i][j]==2-x[j][i])

    m.optimize()


    print('Model status:', m.status)


    for v in m.getVars():
        print(v.varName, v.x)


    print('Obj:', m.objVal)

except GurobiError:
    print('Error reported')
```
